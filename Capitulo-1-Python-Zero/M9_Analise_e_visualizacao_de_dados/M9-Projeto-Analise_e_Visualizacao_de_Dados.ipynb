```python
#2-Qual é o tempo médio, máximo e mínimo de resposta?
print("O tempo medio/mean consultando a func count: 6.667572")
print("O tempo maximo/max consultando a func count: 13.000000")
print("O tempo maximo/max consultando a func count: 0.000000")
#data_frame.describe()
#OU
print("")

print("O tempo média usando func especifica:",data_frame["Tempo Resposta"].mean())
print("O tempo máximo usando func especifica:",data_frame["Tempo Resposta"].max())
print("O tempo mínimo usando func especifica:",data_frame["Tempo Resposta"].min())
```

```python
#3-Qual é a nota média, máxima e mínima do consumidor?
print("Nota do Consumidor média :",data_frame["Nota do Consumidor"].mean())
print("Nota do Consumidor máximo :",data_frame["Nota do Consumidor"].max())
print("Nota do Consumidor mínimo :",data_frame["Nota do Consumidor"].min())
```

```python
#4-Como podemos correlacionar a nota do consumidor com o tempo de resposta? Explique.
plt.figure(figsize=(18,8))

sns.heatmap(data_frame.corr(method = 'pearson'), annot=True, fmt=".1f");

print("""
Podemos observar no head map (feito com metodo de person), que a relação do tempo de resposta é de -0.2 . 
Oque indica uma reta decrescente, ou seja quanto menor o tempo de respota maior a nota. 
Olhando os graficos de barra, podemos ver que temos uma alta quantidade de notas baixas (muitas avaliações ruins), 
e alta quantidade de tempo de resposta alto (demora para responder). É necessario melhorar o setor de resposta ao cliente
""")
```

```python
#-5Qual a quantidade de reclamações por Sexo?
print("Quantidade de reclamação por sexo M:",data_frame[data_frame["Sexo"]=='M'].count()[0])
print("Quantidade de reclamação por sexo F:",data_frame[data_frame["Sexo"]=='F'].count()[0])

print("\n ^ Esta logica acima me parece ser possivel apenar por não haver null na coluna Sexo, a outra a baixo me parece mais apropriada\n")

print(data_frame["Sexo"].value_counts())
```

```python
#-6Qual a quantidade de reclamações por Estado?
data_frame["UF"].value_counts() 
```

```python
#-7Qual é a porcentagem de reclamações registradas e não respondidas?
nao_resolvida = data_frame["Avaliação Reclamação"].eq("Não Resolvida").sum()

porcentagem_nao_resolvida = (nao_resolvida * 100)/43987

a='"'

print(f"Porcentagem de reclamações {a}Não Resolvidas{a} =",porcentagem_nao_resolvida,"%")
```

```python
#5-Gere um gráfico com titulo, nome dos eixos, cor e legenda para as seguintes situações:
#a) Frequência de reclamações por sexo
banana = data_frame[["Avaliação Reclamação","Assunto","Sexo"]].groupby(by=["Avaliação Reclamação","Sexo"]).count()

banana.plot.bar()

plt.title("Frequência de reclamações por sexo")
plt.xlabel("")
plt.show()
```

```python
#5-Gere um gráfico com titulo, nome dos eixos, cor e legenda para as seguintes situações:
#b) Frequencia de reclamações por estado
ar_uf = data_frame[["Avaliação Reclamação","Assunto","UF"]].groupby(by=["Avaliação Reclamação","UF"]).count()

ar_uf.plot.bar()

plt.title("Frequência de reclamações por sexo")
plt.xlabel("")
plt.show()
#Não encontrei forma de deixar o eixo x legivel, se conhecer alguma pf me mostre, obrigado desde já
```

```python
#5-Gere um gráfico com titulo, nome dos eixos, cor e legenda para as seguintes situações:
#c) Frequência de reclamações respondidas e não respondidas
print(data_frame["Avaliação Reclamação"].value_counts())
```

```python
#6-Imagine que você é um Cientista de Dados responsável por medir a satisfação dos clientes.
#Para que você possa surpreender seu chefe, você deve fazer duas análises importantes sobre o data frame, 
#incluindo no mínimo dois gráficos com suas devidas personalizações de titulo, nome dos eixos, legenda, estilos etc.
#Considere que, ao trazer informações relevantes para a empresa em que trabalha,
#mais chances de ter o seu trabalho reconhecido e de ser um profissional com maior valor no mercado. 
#Capriche nos gráficos!
x = data_frame["Tempo Resposta"]
y = data_frame["Nota do Consumidor"]

plt.bar(x, y, color = "plum", label = "Notas")

plt.title("Relação Tempo de resposta, Nota do consumidor")
plt.xlabel("Tempo de resposta")
plt.ylabel("Nota")
plt.legend()
plt.show()
```
