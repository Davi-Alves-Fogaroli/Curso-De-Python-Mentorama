JOIN diagrams! Let‚Äôs look at CROSS JOIN first, because all other JOIN types can be derived from CROSS JOIN:



Lembre-se, em uma jun√ß√£o cruzada (em SQL tamb√©m escrita com uma lista de tabelas separada por v√≠rgulas, historicamente) √© apenas pegar cada item do lado esquerdo e combin√°-lo com cada item do lado direito. Quando voc√™ CROSS JOIN uma tabela de 3 linhas com uma tabela de 4 linhas, voc√™ obter√° 3 √ó 4 = 12 linhas de resultado. Veja, estou usando um caractere ‚Äúx‚Äù para escrever a multiplica√ß√£o. Ou seja, atrav√©s".



Em texto simples, um INNER JOIN √© um CROSS JOIN em que apenas as combina√ß√µes que preenchem um determinado predicado s√£o retidas. Por exemplo





```python
sql = """SELECT
    estudantes.nome_estudante,
    departamentos.nome_departamento
FROM departamentos
INNER JOIN estudantes ON estudantes.id_departamento = departamentos.id_departamento """
#ou
#INNER JOIN estudantes USING(id_departamento)"""
cursor.execute(sql)
cursor.fetchall()
```

```python
sql = """SELECT
    estudantes.nome_estudante,
    departamentos.nome_departamento
    FROM estudantes 
    LEFT JOIN departamentos USING (id_departamento)"""
#ou
#LEFT OUTER JOIN departamentos ON estudantes.id_departamento = departamentos.id_departamento""" 
#LEFT JOIN

cursor.execute(sql)
cursor.fetchall()
```
