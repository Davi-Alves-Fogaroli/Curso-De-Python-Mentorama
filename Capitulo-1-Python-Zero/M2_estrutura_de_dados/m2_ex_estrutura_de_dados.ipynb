    1)Crie um progrma que leia  nome, sexo e idade de varias pessoas, guardando os dados de cada pessoa em um dicionario e todos os dicionario em uma lista.
        a) Quantas pessoas foram cadastradas? x
        b) A média de idade? x
        c) Uma lista com as mulheres
        d) Uma lista de pessoas com idade acima da média

```python
#-----FUNÇÕES-----#
    
def dicionarios(): #esta antes das variaveis pois normalmente estaria em outro arquivo dentro da mesma pasta
    pessoa = {
        "Nome": input("\nDigite seu nome: "),
        "Idade": int(input("Digite sua idade: ")),
        "Sexo": input("Digite seu sexo.(M)mulher, (H)homem, (N)Não Binario: ").upper()
    }
    return pessoa
    
#-----VARIAVEIS-----#
    
listaArmazem=[]
idades=[]
listaMulheres=[]
nomeIdadeMaiorMedia=[]
pessoasCadastradas=0
numeroIdades=0
media=0
cond='S'
    
#-----CADASTRAR PESSOAS-----#
    
while (cond=='S'):
    pessoasCadastradas += 1
    
    listaArmazem.append(dicionarios())
        
    cond = input("\nDeseja inserir mais uma pessoa ? S/N").upper()

#(entrarNosDicionarios)#
for dic in listaArmazem:
    for chave, valor in dic.items():
        
        #(guardarIdades/FazerMedia)#
        if (chave == "Idade"):
            idades.append(valor)        
            
#(media)#      
for x in idades:
    numeroIdades += 1

media=sum(idades)/numeroIdades 

#(entrarNosDicionarios)#        
for dic in listaArmazem:
    for chave, valor in dic.items():
        
        #(pegarNomeDasMulheres)#
        if (chave == "Sexo" and valor == "M"):
            for chave, valor in dic.items():
                if (chave == "Nome"):
                    listaMulheres.append(valor)       
        
        #(pegarNomeIdade>Media)#
        if (chave == "Idade" and valor > media):
            for chave, valor in dic.items():
                if (chave == "Nome"):
                    nomeIdadeMaiorMedia.append(valor)

#-----PRINT-----#
    
print("\n\nNumero de pessoas cadastradas:%s "%pessoasCadastradas)
print("\nMedia de idade:%f "%media)
print("\nNome das mulheres cadastradas: ",listaMulheres)
print("\nAs pessoas com idade maior que a media: ",nomeIdadeMaiorMedia)
```

```python
#EX (9)
def ordena_lista (lista):
    output = []
    numeros = []
    for item in lista:
        if type(item) == str:
            output.append(item)
        elif type(item) == int:
            numeros.append(item)
    numeros.reverse()
    output = output + numeros
    return output

lista = ['A', 1, 'E', 5, 'T', 7, 'W', 8, 'G']

print(ordena_lista(lista))
```
