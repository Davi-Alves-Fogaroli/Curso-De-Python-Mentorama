3) Problema Produtor Consumidor
(A)
-PROBLEMA- É um problema classico de SO que consiste em um empasse, onde por exemplo 2 ou mais processos(P1, P2, Pn)  precisam ter acesso a um recurso(R, que so pode ser utilizado por um P de cada vez), no entanto P1 chegou antes e esta utilizando R, por consequencia P2 ficara em Espera Ociosa, onde ele consome procesamento em um Loop Ocioso (fica sendo escalonado pelo SO até que  consiga usar o R) enquato espera o liberamento do recurso R, para que possa executar sua tarefa.

-SOLUÇÃO- Para ilustra o problema, teremos um produtor(P1), um consumidor(P2), e um armazem(recurso,R) com capacidade de n fardos de alfafa, utilizaremos conseito de ("semafaro, acordado(up) e dormirndo(down) "<-propostos por "Dijkstra"), em resumo a ideia é que se o P1 estive usando o R, faremos o P2 dormir(down) para que não fique em espera ociosa.
Utilizando 3 semafaros, 2 contadores e 1 binario.

Contadores: full e empty, responsaveis por sincronizar os processor, não permitir que o produtor produza quando o buff estiver cheio, nem que o consumidor consuma quando ele estiver vazio.

Binario: Conhecido tambem como "MUTEX"(mutual exclusion) tem função de garantir que somento um P por vez use um determinado recurso em determinado instante de tempo. Em nossa solução ele é o gerenciador do armazér(R), não permitindo que P1 e P2 o utilizem ao mesmo tempo, quando NÃO esta sendo usado ele possui "valor = 1", até que seja "desocupado", e fique com "valor = 0" quando ESTA sendo utilizado.

(B) Problema dos filozofos no jantar
Cinco filósofos sentados ao redor de uma mesa redonda, cada qual fazendo exclusivamente uma das duas coisas: comendo ou pensando. Enquanto está comendo, um filósofo não pode pensar, e vice-versa.
Cada filósofo possui um prato cheio de spaghetti à sua frente. Além disso, um garfo é posicionado entre cada par adjacente de filósofos (portanto, cada filósofo tem exatamente um garfo à sua esquerda e exatamente um garfo à sua direita). Assume-se que um filósofo necessita de dois garfos para poder comer. E, além disso, um filósofo só pode utilizar um garfo que esteja imediatamente à sua esquerda ou imediatamente à sua direita.
Um filósofo só pode mudar para o estado 'comendo' se nenhum dos vizinhos estiver comendo. Os vizinhos do filósofo i são definidos pelas macros LEFT e RIGHT. Em outras palavras, se i for 2, LEFT será 1 e RIGHT será 3.
O programa usa um arranjo de semáforos, um por filósofo; assim, filósofos famintos podem ser bloqueados se os garfos necessários estiverem ocupados. 
A diferença entre este e o outro problrma é a quantidade de recursos disponiveis, que acarreta em mais de uma Thread funcionando ao mesmo tempo(usando recursos diferentes claro, por conta do MUTEX/LOCK).
