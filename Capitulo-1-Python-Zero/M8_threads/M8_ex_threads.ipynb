3) Problema Produtor Consumidor
(A)
-PROBLEMA-  um problema classico de SO que consiste em um empasse, onde por exemplo 2 ou mais processos(P1, P2, Pn)  precisam ter acesso a um recurso(R, que so pode ser utilizado por um P de cada vez), no entanto P1 chegou antes e esta utilizando R, por consequencia P2 ficara em Espera Ociosa, onde ele consome procesamento em um Loop Ocioso (fica sendo escalonado pelo SO at que  consiga usar o R) enquato espera o liberamento do recurso R, para que possa executar sua tarefa.

-SOLUO- Para ilustra o problema, teremos um produtor(P1), um consumidor(P2), e um armazem(recurso,R) com capacidade de n fardos de alfafa, utilizaremos conseito de ("semafaro, acordado(up) e dormirndo(down) "<-propostos por "Dijkstra"), em resumo a ideia  que se o P1 estive usando o R, faremos o P2 dormir(down) para que no fique em espera ociosa.
Utilizando 3 semafaros, 2 contadores e 1 binario.

Contadores: full e empty, responsaveis por sincronizar os processor, no permitir que o produtor produza quando o buff estiver cheio, nem que o consumidor consuma quando ele estiver vazio.

Binario: Conhecido tambem como "MUTEX"(mutual exclusion) tem funo de garantir que somento um P por vez use um determinado recurso em determinado instante de tempo. Em nossa soluo ele  o gerenciador do armazr(R), no permitindo que P1 e P2 o utilizem ao mesmo tempo, quando NO esta sendo usado ele possui "valor = 1", at que seja "desocupado", e fique com "valor = 0" quando ESTA sendo utilizado.

(B) Problema dos filozofos no jantar
Cinco filsofos sentados ao redor de uma mesa redonda, cada qual fazendo exclusivamente uma das duas coisas: comendo ou pensando. Enquanto est comendo, um filsofo no pode pensar, e vice-versa.
Cada filsofo possui um prato cheio de spaghetti  sua frente. Alm disso, um garfo  posicionado entre cada par adjacente de filsofos (portanto, cada filsofo tem exatamente um garfo  sua esquerda e exatamente um garfo  sua direita). Assume-se que um filsofo necessita de dois garfos para poder comer. E, alm disso, um filsofo s pode utilizar um garfo que esteja imediatamente  sua esquerda ou imediatamente  sua direita.
Um filsofo s pode mudar para o estado 'comendo' se nenhum dos vizinhos estiver comendo. Os vizinhos do filsofo i so definidos pelas macros LEFT e RIGHT. Em outras palavras, se i for 2, LEFT ser 1 e RIGHT ser 3.
O programa usa um arranjo de semforos, um por filsofo; assim, filsofos famintos podem ser bloqueados se os garfos necessrios estiverem ocupados. 
A diferena entre este e o outro problrma  a quantidade de recursos disponiveis, que acarreta em mais de uma Thread funcionando ao mesmo tempo(usando recursos diferentes claro, por conta do MUTEX/LOCK).
