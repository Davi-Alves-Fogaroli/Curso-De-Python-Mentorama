```python
#EX1 M4 Modo 1

import socket #socket me permitem utilizar redes

invalidos=[]
enderecos=["200.135.80.9","192.168.1.1","8.35.67.74","257.32.4.5","85.345.1.2","1.2.3.4","9.8.234.5","192.168.0.256"]

for ip in enderecos:
    try:
        socket.inet_aton(ip)#inet_aton() essa função converte
    except:
        invalidos.append(ip)

validos = set(enderecos) - set(invalidos)

print("Validos:",validos)
print("\nInvalidos:",invalidos)
```

B)Erro de tempo de execução: Esse erro pode ser um problema tanto em algum software, hardware e até um proplema de rede. Por exemplo quando uma requisição demora de mais para ser enviada ou recebida devido a uma falha no banco de dados, uma queda de rede durante a comunicação com a API. Podem ser causados em um site que usa código HTML que não é compatível com a funcionalidade do navegador da Web.

C)Erro de semântica: É um erro onde um comando ou instrução esta mao escrito, as funções foram passadas corretamente sem nenhum erro de sintax, no entanto a instrução nçao faz o menor sentido, como por exemplo:
-Uma divisão por uma string      "12/dois"
-Uma divisão de     "0/0"
-Uma classe herdar de si mesma
-Por uma parte de uma função dentro de uma outra intrução   "for if in range(match)"
Erros de semântica podem ou não serem pegos pelo debugin, e poder gerar erros de tempo de execução.
