VAMOS TRANSFOMAR DUAS LISTAS EM UM DICIONARIO, COM CHAVE E VALOR


```python
lista_nomes_futuras_chaves = ["José", "Claudia", "Joares"]
lista_notas_futuros_valores = [4, 7, 9.6] 

print(lista_nomes_futuras_chaves, "\n", lista_notas_futuros_valores)

```



ELA VAI PEGAR O PRIMEIRO ELEMENTO DA PRIMEIRA LISTA E JUNTAR AO PRIMEIR ELEMENTIO DA SEGUNDA LISTA


```python
#Criaremos uma lista de nomes e notar para receber o resultado dessa função zip
lista_nomes_notas = list(zip(lista_nomes_futuras_chaves, lista_notas_futuros_valores))

#essa função ao juntar os paraes fez tuplas, que eu guardeu em um objeto de lista
print(lista_nomes_notas)
```

```python
dicionario = dict(lista_nomes_notas)
print(dicionario)
```
