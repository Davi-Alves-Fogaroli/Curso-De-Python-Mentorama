```python
from time import perf_counter

print("insira 2 valores de uma só vez com espaço entre eles\n o primeiro coresponde a vezes de execução da divição do numero futuramente pedido \n o segundo é o valor que executara a divição do numero futuramente pedido ")
n, m = map(int, input().split()) #map() ???????

tempo1_comeca = perf_counter()

for i in range(n): #usuario deu entrada do valor n
    print("insira valor inteiro")
    t = int(input())
    if t % m == 0:
        print(f"Valor inserido {t} é divisivel pelo valor: {m}")
        
tempo1_termina = perf_counter()

print("tempo de execução: ", tempo1_termina-tempo1_comeca)
```

```python
import datetime

day_delta = datetime.timedelta(days=1)

comeco_data = datetime.date.today()
fim_data = comeco_data + 5*day_delta

for i in range((fim_data - comeco_data).days):
    print(comeco_data + i*day_delta)
```
