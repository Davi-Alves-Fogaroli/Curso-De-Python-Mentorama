```python
from time import perf_counter
import time

comeco = perf_counter()
comeco_time = time.time()

for i in range(0,3,1):
    if (i == 0):
        print("Hola mundo")
    time.sleep(3)
    print("Hola Mundo")
    
fim = perf_counter()
fim_time = time.time()

print(f"\nEste codigo demorou {fim-comeco} segundos para ser executado")
print("Usando Time.time():",fim_time-comeco_time)
```

```python
#  4  (VERSO SIMPLES)

import datetime

passo_delta = datetime.timedelta(days=1)

data_inicial = datetime.date(1987,10,16)

data_final = datetime.date(1987,10,25)

#data_ajustada == data_ajustada_para_intervaulo_sem_considerar_a_data_inicial 
data_ajustada = data_inicial + passo_delta

for i in range((data_final - data_ajustada).days):
    print((data_inicial + passo_delta) + i * passo_delta)
```

```python
# 4  (VERSO COM ARMAZENAMENTO)

import datetime

lista = []

passo_delta = datetime.timedelta(days=1)

data_inicial = datetime.date(1987,10,16)

data_final = datetime.date(1987,10,25)

#data_ajustada == data_ajustada_para_intervaulo_sem_considerar_a_data_inicial 
data_ajustada = data_inicial + passo_delta

for i in range((data_final - data_ajustada).days):
    resultado = 0
    resultado = ((data_inicial + passo_delta) + i * passo_delta)
    lista.append(resultado)

print(lista)
```

```python
# 6
from datetime import datetime

def converter_string_mes_em_data(string1):

    if 'janeiro' in string1:
        string1 = string1.replace('janeiro', '01')

    if 'fevereiro' in string1:
        string1 = string1.replace('fevereiro', '02')

    if 'maro' in string1:
        string1 = string1.replace('maro', '03')

    if 'abril' in string1:
        string1 = string1.replace('abril', '04')

    if 'maio' in string1:
        string1 = string1.replace('maio', '05')

    if 'junho' in string1:
        string1 = string1.replace('junho', '06')

    if 'julho' in string1:
        string1 = string1.replace('julho', '07')

    if 'agosto' in string1:
        string1 = string1.replace('agosto', '08')

    if 'setembro' in string1:
        string1 = string1.replace('setembro', '09')

    if 'outubro' in string1:
        string1 = string1.replace('outubro', '10')

    if 'novembro' in string1:
        string1 = string1.replace('novembro', '11')

    if 'dezembro' in string1:
        string1 = string1.replace('dezembro', '12')
    
    return string1

string1 = converter_string_mes_em_data('20 janeiro 2021')

data1 = datetime.strptime(string1, '%d  %m  %Y').date()

print(data1)
```
