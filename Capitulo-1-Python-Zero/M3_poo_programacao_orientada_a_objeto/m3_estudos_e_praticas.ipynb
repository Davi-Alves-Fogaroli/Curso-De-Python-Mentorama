            4 POLIMORFISMO

```python
#criando a classe FromaGeometrica
class FormaGeometrica():
    def calculo_area(self):
        pass
```

```python
#Criando as class quadrado e circulo
class Quadrado(FormaGeometrica):
    def __init__(self, base): #metodo origial da classe
        self.base = base
        
    def calculo_area(self): #metodo herdado que esta sofrendo POLIMORFISMO 
        return self.base * self.base
    
class Circulo(FormaGeometrica):
    def __init__(self, raio): #__init__ construtor ???
        self.raio = raio
        
    def calculo_area(self):
        return self.raio * self.raio * 3.14
```

```python
#Utiliando os metodos polimorfizados

quadrado1 = Quadrado(2)
quadrado2 = Quadrado(3)

circulo1 = Circulo(1)
circulo2 = Circulo(2)

formas_geometricas = [quadrado1, quadrado2, circulo1, circulo2]

soma = 0

for forma in formas_geometricas:
    soma = soma + forma.calculo_area()

    print("A soma das areas √© {}".format(soma))
```
