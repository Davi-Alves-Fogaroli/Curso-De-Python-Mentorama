```python
#ex(2)#
class Quadrado():
    def __init__(self, tamanho_do_lado):
        self.tamanho_do_lado = tamanho_do_lado

    def mudar_valor_do_lado(self, novo_tamanho):
        self.tamanho_do_lado = novo_tamanho
    
    def retornar_valor_do_lado(self):
        return self.tamanho_do_lado
    
    def calcular_area(self):
        return self.tamanho_do_lado * self.tamanho_do_lado

quadrado1 = Quadrado(2)   

print("Valor definido do quadrado1 logo na sua criação = ", quadrado1.tamanho_do_lado)

quadrado1.mudar_valor_do_lado(4)

print("Valor do qadrado1, pos modificação = ", quadrado1.retornar_valor_do_lado())

print("Valor da area do quadrado1 = ",quadrado1.calcular_area())
```

```python
#__init__ é um metodo resevado em python POO, 
#é chamado/usado quando um obj é criado apartir de uma classe e permite a classe inicializar os atributos da classe
#self utilizado para acessar as instancias de uma classe, ou seja, atributos e metodos

#ex(3)#
class Retangulo():
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
        
    def mudar_valor_dos_lados(self, nova_base, nova_altura):
        self.base = nova_base
        self.altura = nova_altura
    
    def mostrar_valor_dos_lados(self):
        return self.base, self.altura
    
    def calcular_area(self):
        return self.base * self.altura 
    
    def calcular_perimetro(self):
        return (self.base *2) + (self.altura*2)

    
retangulo1 = Retangulo(int(input("Declare base do retangulo: ")),int(input("Declare altura do retangulo: "))) 
 
print("\nOs lados do seu retangulo são:",retangulo1.mostrar_valor_dos_lados())

print("Serão necessarios =", retangulo1.calcular_area(), "m^2 de piso")

print("Serão necessarios =", retangulo1.calcular_perimetro(), "m de roda pé")
```

```python
#__init__ usado, chamado quando uma obj é criado aparir de uma classe e permite, a classe inicializar os atributos da classe
#self permite acessar as instacias(atributos e metodos efinidos) da classe, os atributos e metodos

#ex(4)#
class Pessoa():
    def __init__(self, nome , idade, peso, altura):
        self.nome = nome
        self.idade = idade
        self.peso = peso
        self.altura = altura
        
    def engordar(self, ganho_de_peso):        
        self.peso += ganho_de_peso
        return self.peso
    
    def emagrecer(self, perder_peso):        
        self.peso -= perder_peso
        return self.peso
    
#se idade menor que 21, a cada ano passado, acrescenta 0.5cm
#if idade < 21 :  for x in range (idade+1ou nao ?): tamanho += 0.5 
    def envelhecer(self, idade):
        self.idade += idade
        if (self.idade < 21):
            for x in range(idade):
                self.altura += 0.05
        
    def crescer(self, cresceu):
        self.altura += cresceu
        return self.altura
    
    def status(self):
        print("Status\nnome:",self.nome)
        print("idade:",self.idade)
        print("pesso:",self.peso)
        print("altura:",self.altura)
        
    
pessoa1 = Pessoa("Davi", 14, 53.7, 1)

pessoa1.status()
pessoa1.envelhecer(3)
print("\nIdade depois da soma:",pessoa1.idade, "Altura depois da soma:",pessoa1.altura)
print("\nPeso antes de engordar:",pessoa1.peso,"Engordou e ficou com:",pessoa1.engordar(1),"Kg")
print("\nPeso depois de engordar:",pessoa1.peso,"Emagreceu e ficou com:",pessoa1.emagrecer(2),"Kg")
print("\nAltura antes de crescer de noite hehe:",pessoa1.altura,"Depois de crescer:",pessoa1.crescer(.1))
```

```python
# 3 atributos mim recebam entrada
# 3 metodos mim  mecham com os atributos
# chamar os atributos
#__init__ é chamdo quando um obj é criado atraves de uma class, e permite a classe inicializar os atribudos da classe
#self permite acessa as instancias(atributos e metodos) de uma classe, (set e get do java hehe) 

"""
OBS IMPORTANTE, ao freiar continua desacelerando, para acelerar é necessario superar a desaceleração, logo o valor 
da aceleração e tende ser maior que o da desaceleração pra que o veiculo pare de desacelerar.
"""

#ex(5)#
class Carro():
    def __init__(self, acelerar, velocidade, freiar):
        self.acelerar = acelerar
        self.velocidade = velocidade
        self.freiar = freiar
    
    def aceleracao(self, acelera):
        self.acelerar += (acelera - self.freiar)
        self.velocidade += self.acelerar
        self.freiar -= acelera
        
        if (self.freiar < 0):
            self.freiar = 0
        
    def freiando(self, freiando):
        self.freiar += (freiando - self.acelerar)
        self.acelerar -= freiando
        self.velocidade -= self.freiar
        
        if (self.acelerar < 0):
            self.acelerar = 0
        elif (self.velocidade < 0):
            self.velocidade = 0
        
    def mostraSeEstaAcimaDoLimite(self):
        if (self.velocidade >= 80):
            print("Voçê esta acima do limite permitido, sua velocidade é de:", self.velocidade, ",o permitido é 80km/h" )
    
    def dados(self):
        print("\nDados atuais do carro:", "\nAceleração:", self.acelerar, "\nVelocidade:", self.velocidade, "\nFrenagem:", self.freiar)
    
carro1 = Carro(0, 90, 0)


carro1.freiando(7)
carro1.aceleracao(8)

carro1.mostraSeEstaAcimaDoLimite()

carro1.dados()
```

```python
#ex (6)#

class ContaCorrente():
    
    def __init__(self, numeroDaConta, nomeDoCorrentista, saldo):
        self.numeroDaConta = numeroDaConta
        self.nomeDoCorrentista = nomeDoCorrentista
        self.saldo = saldo
        
    def alterarNome(self, novoNome):
        self.nomeDoCorrentista = novoNome
        return self.nomeDoCorrentista
    
    def deposito(self, depositar):
        self.saldo += depositar
        print("\nForan depositados R$:", depositar, "\tSeu saldo atual é de R$:",self.saldo)
    
    def saque(self, sacar):
        self.saldo -= sacar
        print("\nForan sacados R$:", sacar, "\t\tSeu saldo atual é de R$:",self.saldo)
    
conta1 = ContaCorrente(37521, "josé", 0)

print(f"O novo titular da conta {conta1.numeroDaConta} se chama:",conta1.alterarNome("Julio"))
conta1.deposito(57)
conta1.saque(157)
```

```python
#__init__ metodo construtor que cria um obj atraves da class, e permite a classe inicializar os atributos dentro da classe
#self permite acessar os metodos e atributos (instancias) de uma class

#ex(7)#  

"""Volume maximo = 20
Volume minimo = 0"""

class Tv():
    def __init__(self, numeroCanal, volume):
        self.numeroCanal = numeroCanal
        self.volume = volume
        
    def aumentarVolume(self, almentandoVolume):
        self.volume += almentandoVolume
        if (self.volume > 20):
            self.volume = 20
        print("Volume:",self.volume)
            
    def diminuirVolume(self, diminuindoVolume):
        self.volume -= diminuindoVolume
        if (self.volume < 0):
            self.volume = 0            
        print("Volume:", self.volume)
        
    def mudarCanal(self, mudandoCanal):
        self.numeroCanal = mudandoCanal
        print("Canal:", self.numeroCanal)

televisor1 = Tv(2, 20)

televisor1.aumentarVolume(2)
televisor1.diminuirVolume(30)
televisor1.mudarCanal(3)
```

```python
#ex(8)#

class BombaCombustivel():
    def __init__(self, tipoCombustivel, custoLitro, quantidadeCombustivel):
        self.tipoCombustivel = tipoCombustivel
        self.custoLitro = custoLitro
        self.quantidadeCombustivel = quantidadeCombustivel
        
    def abastecerPorValor(self, valorDoAbastecimento):
        #5=1litro
        litrosAbast = valorDoAbastecimento/5
        self.quantidadeCombustivel -= litrosAbast
        if (self.quantidadeCombustivel < 0):
            self.quantidadeCombustivel += litrosAbast
            print("\nNão ha combustivel suficiente para o abastecimento, combustivel disponivel:",self.quantidadeCombustivel)
        else:
            print("\nForam abatecidos:",litrosAbast,"litros")
    
    def abastecerPorLitro(self, litrosAbastecido):
        #1=5r$
        valorDoAbastecimento = litrosAbastecido*5
        self.quantidadeCombustivel -= litrosAbastecido
        if (self.quantidadeCombustivel < 0):
            self.quantidadeCombustivel += litrosAbastecido
            print("\nNão ha combustivel suficiente para o abastecimento, combustivel disponivel:",self.quantidadeCombustivel)
        else:
            print("\nCusto do abastecimento R$",valorDoAbastecimento)
        
    def mudaCustoDoLitro(self, novoCusto):
        self.custoLitro = novoCusto
        print("\nNovo custo litro R$", self.custoLitro)
        
    def mudarCombustivel(self, novoCombustivel):
        self.tipoCombustivel = novoCombustivel
        print("\nCombusivel trocado, novo tipo:", self.tipoCombustivel)
        
    def mudarQuantidadeCombustivel(self, novaQuantidade):
        self.quantidadeCombustivel = novaQuantidade
        print("\nO combustivel da bomba teve sua quantidade manipulada, nova quantidade:", self.quantidadeCombustivel)
        
bomba1 = BombaCombustivel("Gasolina", 5, 15)

bomba1.abastecerPorValor(50)
bomba1.abastecerPorLitro(10)
bomba1.mudarCombustivel("Etanol")
bomba1.mudaCustoDoLitro(10)
bomba1.mudarQuantidadeCombustivel(300)
bomba1.abastecerPorLitro(10)
```
