```python
# imports
from bs4 import BeautifulSoup #imporando a sopa bonita, para fazermos nossos scraping, crawler...
from requests import get #importando funÃ§Ã£o para fazer requisiÃ§Ã£o a api
import html5lib #importa um parset bom para trabalhar com html5
import re #Regular Expression >>> Tem por objetivo persquisar e alterar textos usando padrÃµes formais.

# requisitanto a informaÃ§Ã£o do link e armazenando em uma variavel
site = get("https://en.wikipedia.org/wiki/Call_of_Duty")

# "baixando" o html armazenado na variavel "site" e atribuindo a um obj do tipo BeautifulSoup (<class 'bs4.BeautifulSoup'>)
soup = BeautifulSoup(site.content, "html5lib")
##print(type(soup))

# procurando o titulo do html armazenado em "soup" e guardando-o na variavel
titulo = soup.find("title")

print(f"PÃ¡gina principal: >{titulo.text}")

# procurando dentro do html armazenado em "soup", tudo que tem a teag "body", e armazenando na var "texto"
texto = soup.find_all("body")

# percorrendo a variavel que recebeu o html do body
for f in texto:
    # procurando dentro de cada percorredor 'f' todos os links, e filtrando com auxilio da biblioteca "Regular Expression" 
    todos_links = f.find_all('a', href = re.compile("/wiki/"))
    
    # percorendo a variavel que recebe os links encontrados na(s) tag "body"
    for link in todos_links:
        # pegando o titulo de cada link do percorredor "link" 
        titulo_segundario = link.get("title")
        
        print(f"Pagina segundaria >> {titulo_segundario}")
```
