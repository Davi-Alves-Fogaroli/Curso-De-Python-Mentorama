```python
# tenha instalado
#-bs4 (para usar bealtifulSoup)
#-html5lib (para usar o parset)

# importando
from requests import get
from bs4 import BeautifulSoup
import html5lib
import re

response = get("https://pt.wikipedia.org/wiki/Call_of_Duty:_World_at_War:_Zombies")

soup = BeautifulSoup(response.text, "html5lib")

title = soup.find("title")

print(f"Página principal:{title.text}")

corpo = soup.find_all("body")

for a in corpo:
    links = a.find_all('a', href = re.compile("/wiki/"))
    
    for r in links:
        titulo_2 = r.get("title")
    
        print("t2: ",titulo_2)
```
