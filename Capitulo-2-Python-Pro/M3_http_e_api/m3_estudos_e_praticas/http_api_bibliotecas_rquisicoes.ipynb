

#### HTTP:  HyperText Transfer Protocol
    - uilizado na transferencia de texto
    -  um protocolo de transferenci de hiper-texto(so capases de fazer ligaes entr texto, so textos com estrutura no linear)
    - serve ara ler texto na web
    - HTTP no  s um protocolo para enviar uma requisio a um site
    - Eles fazem o protocolo de toda a comunicao,a resposta que o servidor envia faz parte do protocolo (esses protocolos padronizam e ajudam nas tarefa)
    - Existe a muito tempo, mais de 1980
    
#### HTTPS:  Hyper Text Transfer Protocol Secure
    - utilizado na transferencia de arquivo ( um protocolo)
    - HTTPS no  s um protocolo para enviar uma requisio a um site 
    - Eles fazem o protocolo de toda a comunicao,a resposta que o servidor envia faz parte do protocolo (esses protocolos padronizam e ajudam nas tarefa)
    - Os dados so encripitados diferentemente do HTTP
    
##### Metodos: get e set (servem para estabelecer a comunicao entre cliente e servidor)
