

#### HTTP:  HyperText Transfer Protocol
    - uilizado na transferencia de texto
    - é um protocolo de transferenci de hiper-texto(são capases de fazer ligações entr texto, são textos com estrutura não linear)
    - serve ara ler texto na web
    - HTTP não é só um protocolo para enviar uma requisição a um site
    - Eles fazem o protocolo de toda a comunicação,a resposta que o servidor envia faz parte do protocolo (esses protocolos padronizam e ajudam nas tarefa)
    - Existe a muito tempo, mais de 1980
    
#### HTTPS:  Hyper Text Transfer Protocol Secure
    - utilizado na transferencia de arquivo (é um protocolo)
    - HTTPS não é só um protocolo para enviar uma requisição a um site 
    - Eles fazem o protocolo de toda a comunicação,a resposta que o servidor envia faz parte do protocolo (esses protocolos padronizam e ajudam nas tarefa)
    - Os dados são encripitados diferentemente do HTTP
    
##### Metodos: get e set (servem para estabelecer a comunicação entre cliente e servidor)
